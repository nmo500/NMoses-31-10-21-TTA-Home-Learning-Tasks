```python
# PYTHON TWO, TASK ONE - IS THE NUMBER GREATER THAN OR LESS THAN 10?

 #Task 1: Write a program that asks the user for 
#an integer number and checks if it is > 10. If it 
#is, it will print “Number is Greater than 10”, 
#else “Number is smaller than 10”.

inputNumberToSeeIfGreaterThan10 = input("Write a number and I will tell you if it is greater than or less than 10! ")
inputNumberToSeeIfGreaterThan10 = float (inputNumberToSeeIfGreaterThan10)
if inputNumberToSeeIfGreaterThan10 >10:
  print ("That number is BIGGER than 10!")
elif inputNumberToSeeIfGreaterThan10 <10:
    print ("That number is SMALLER than 10!")

else :
    print ("You have chosen the number 10 - that is EQUAL to 10!")
```

```python
# PYTHON TWO, TASK TWO - USING A LOOP TO ADD TO A VALUE UNTIL IT REACHES 10

#Write a loop program that ask the 
#user for an integer number and check if it is < 
#10. If it is < 10 then it keeps adding 1 to the 
#value.

input_integer_number = input ("Please input a number")
input_integer_number = int (input_integer_number)


current_value = (input_integer_number)
while current_value <= 10:
 print(current_value)
 current_value += 1

if current_value := 10:
  print ("The number has now reached 10")
```
